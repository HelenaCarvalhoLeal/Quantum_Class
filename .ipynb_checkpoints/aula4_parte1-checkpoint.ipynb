//simulações de circuito e implementação em computadores reais


```python
import numpy as np
from qiskit import QuantumCircuit
from qiskit.providers.basic_provider import BasicSimulator
from qiskit.visualization import plot_histogram, plot_distribution
```

```python
simulador = BasicSimulator()
```

```python
circuit = QuantumCircuit(2,2) #2 qubits e 2 bits
circuit.h(0) #hadamard
circuit.measure([0,1],[0,1]) #medir o qibit 1, registarndo no 1
circuit.draw("mpl")
```

```python
job = simulador.run(circuit, shots=10, memory=True)
result = job.result()
#print(result) #só para ficar mais clean
counts = result.get_counts()
print(counts)
```

```python
#plot_distribution(counts)
plot_histogram(counts)
```

```python
result.get_memory()
```

```python
qc = QuantumCircuit(2,2)
qc.rx(np.pi/3,0)
qc.measure(range(2), range(2))
qc.draw("mpl")
```

```python
from qiskit import transpile
qc_t = transpile(qc,backend=simulador)
qc_t.draw("mpl")
```

```python
job = simulador.run(qc_t, shots=1000)
result = job.result()
counts = result.get_counts()
print(counts)
```

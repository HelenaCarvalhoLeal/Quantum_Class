### 2.54) Calcule a energia de estado fundamental do oscilador harmônico para cinco algarismos significativos pelo método do cachorro sacudindo o rabo. Isto é, resolva a Equação 2.72 numericamente, variando K até que você obtenha uma função de onda que vá a zero em ξ grande. 

#### "Na Mathematica, o código de entrada apropriado seria Plot[Evaluate[u[x]/.NDSolve[fu"{xl -(x2 - K)*u[x] == O, u[O] == l,u'[O] == O}, u[x), (x, l0-8, 10), MaxSteps -> 10000]], {x, a, b), PlotRange -> {c, d)];"

##### Vou mostrar apenas os dois primeiros gráficos e os dois últimos. Evidentemente K está entre 0,9999 e 1,0001.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def oscillator_ode(x, u, K):
    return [u[1], (x**2 - K) * u[0]]

def plot_wave_function(K, a=0, b=10, c=-10, d=10):
    u0 = [1, 0]  # u(0) = 1, u'(0) = 0
    sol = solve_ivp(oscillator_ode, [a, b], u0, args=(K,), t_eval=np.linspace(a, b, 1000))
    
    plt.plot(sol.t, sol.y[0], label=f'K = {K}')
    plt.ylim(c, d)
    plt.xlabel('x')
    plt.ylabel('u(x)')
    plt.title(f'Oscilador Harmônico para K = {K}')
    plt.legend()
    plt.grid(True)

# Plot para K = 0.9
plt.figure(figsize=(12, 8))
plot_wave_function(0.9)
plt.show()

# Plot para K = 1.1
plt.figure(figsize=(12, 8))
plot_wave_function(1.1)
plt.show()

# Ajustar o intervalo de valores de K
#plt.figure(figsize=(12, 8))
#plot_wave_function(0.95)
#plt.show()

#plt.figure(figsize=(12, 8))
#plot_wave_function(1.05)
#plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Definir a equação diferencial do oscilador harmônico
def oscillator_ode(x, u, K):
    return [u[1], (x**2 - K) * u[0]]

# Função para plotar a função de onda com intervalo e limites ajustados
def plot_wave_function(K, a=1e-8, b=10, c=-10, d=10, plot_range=(-10, 10)):
    u0 = [1, 0]  # Condições iniciais: u(0) = 1, u'(0) = 0
    sol = solve_ivp(oscillator_ode, [a, b], u0, args=(K,), t_eval=np.linspace(a, b, 1000), max_step=1e-3)
    
    plt.plot(sol.t, sol.y[0], label=f'K = {K}')
    plt.ylim(c, d)
    plt.xlim(4, 5.5)  # Limites no eixo x para observar o comportamento entre x = 4 e x = 5.5
    plt.xlabel('x')
    plt.ylabel('u(x)')
    plt.title(f'Oscilador Harmônico para K = {K}')
    plt.legend()
    plt.grid(True)

# Gráfico para K = 0.9999
plt.figure(figsize=(8, 6))
plot_wave_function(0.9999, c=-10, d=10, plot_range=(-10, 10))  # Ajuste no range do eixo y
plt.show()

# Gráfico para K = 1.0001
plt.figure(figsize=(8, 6))
plot_wave_function(1.0001, c=-10, d=10, plot_range=(-10, 10))  # Ajuste no range do eixo y
plt.show()
```

Na Mathematica, o código de entrada apropriado seria Plot[Evaluate[u[x]/.NDSolve[fu"{xl -(x2 - K)*u[x] ==
O, u[O] == l,u'[O] == O}, u[x), (x, l0-8, 10), MaxSteps -> 10000]], {x, a, b), PlotRange -> {c, d)]; 

Vou mostrar apenas os dois primeiros gráficos e os dois últimos. Evidentemente K está entre 0,9999 e 1,0001.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def oscillator_ode(x, u, K):
    return [u[1], (x**2 - K) * u[0]]

def plot_wave_function(K, a=0, b=10, c=-10, d=10):
    u0 = [1, 0]  # u(0) = 1, u'(0) = 0
    sol = solve_ivp(oscillator_ode, [a, b], u0, args=(K,), t_eval=np.linspace(a, b, 1000))
    
    plt.plot(sol.t, sol.y[0], label=f'K = {K}')
    plt.ylim(c, d)
    plt.xlabel('x')
    plt.ylabel('u(x)')
    plt.title(f'Oscilador Harmônico para K = {K}')
    plt.legend()
    plt.grid(True)

# Plot para K = 0.9
plt.figure(figsize=(12, 8))
plot_wave_function(0.9)
plt.show()

# Plot para K = 1.1
plt.figure(figsize=(12, 8))
plot_wave_function(1.1)
plt.show()

# Ajustar o intervalo de valores de K
#plt.figure(figsize=(12, 8))
#plot_wave_function(0.95)
#plt.show()

#plt.figure(figsize=(12, 8))
#plot_wave_function(1.05)
#plt.show()
```

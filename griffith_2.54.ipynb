Na Mathematica, o código de entrada apropriado seria Plot[Evaluate[u[x]/.NDSolve[fu"{xl -(x2 - K)*u[x] ==

Vou mostrar apenas os dois primeiros gráficos e os dois últimos. Evidentemente K está entre 0,9999 e 1,0001.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def oscillator_ode(x, u, K):
    return [u[1], (x**2 - K) * u[0]]

def plot_wave_function(K, a=0, b=10, c=-10, d=10):
    u0 = [1, 0]  # u(0) = 1, u'(0) = 0
    sol = solve_ivp(oscillator_ode, [a, b], u0, args=(K,), t_eval=np.linspace(a, b, 1000))
    
    plt.plot(sol.t, sol.y[0], label=f'K = {K}')
    plt.ylim(c, d)
    plt.xlabel('x')
    plt.ylabel('u(x)')
    plt.title(f'Oscilador Harmônico para K = {K}')
    plt.legend()
    plt.grid(True)

# Plot para K = 0.9
plt.figure(figsize=(12, 8))
plot_wave_function(0.9)
plt.show()

# Plot para K = 1.1
plt.figure(figsize=(12, 8))
plot_wave_function(1.1)
plt.show()

# Ajustar o intervalo de valores de K
#plt.figure(figsize=(12, 8))
#plot_wave_function(0.95)
#plt.show()

#plt.figure(figsize=(12, 8))
#plot_wave_function(1.05)
#plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Definir a equação diferencial (oscilador harmônico)
def harmonic_oscillator(x, y, K):
    u, u_prime = y
    return [u_prime, (x**2 - K) * u]

# Definir função para resolver a equação diferencial
def solve_oscillator(K, x_range):
    # Condições iniciais
    y0 = [1, 0]  # u(0) = 1, u'(0) = 0
    sol = solve_ivp(harmonic_oscillator, x_range, y0, args=(K,), max_step=1e-3)
    return sol.t, sol.y[0]

# Plotar os resultados para diferentes valores de K
x_range = [1e-8, 10]
K_values = [0.9999, 1.0001]

plt.figure(figsize=(8, 6))

# Plotar para K = 0.9999
x1, u1 = solve_oscillator(K_values[0], x_range)
plt.plot(x1, u1, label=f'K = {K_values[0]}')
plt.xlim(4, 5.5)
plt.ylim(-10, 10)

# Plotar para K = 1.0001
x2, u2 = solve_oscillator(K_values[1], x_range)
plt.plot(x2, u2, label=f'K = {K_values[1]}')
plt.xlim(4, 5.5)
plt.ylim(-10, 10)

plt.xlabel(r'$\xi$')
plt.ylabel(r'$u(\xi)$')
plt.title('Solução numérica do Oscilador Harmônico')
plt.legend()
plt.show()
```

Utilize o Qiskit para construir e simular um circuito quântico com dois qubits, incorporando operações unitárias locais e não-locais. Ilustre através da esfera de Bloch.


```python
from qiskit import QuantumCircuit
from qiskit.visualization import visualize_transition
from qiskit.visualization import plot_bloch_multivector
import numpy as np
```

```python
qc = QuantumCircuit(2)
qc.h(0)  # Hadamard em qubit 0 (local)

qc.x(1)  # NOT em qubit 1 (local)
qc.cx(1,0)  # CNOT com controle em qubit 0 e alvo em qubit 1 (não local)

qc.draw("mpl")
```

```python
plot_bloch_multivector(qc)
```

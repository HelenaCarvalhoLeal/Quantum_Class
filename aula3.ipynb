```python
from qiskit import QuantumCircuit
from qiskit.visualization import visualize_transition
from qiskit.visualization import plot_bloch_multivector
import numpy as np
```

```python
qc = QuantumCircuit(3)
qc.h(0) #hadamard
qc.x(1) #NOT
qc.y(1) #Y
qc.z(0) #Z
qc.cx(0,2)
qc.rx(np.pi/4,2)
qc.ry(np.pi/3,0)
qc.rz(np.pi/8,0)
qc.cz(1,0)
qc.draw("mpl")
```

```python
qc = QuantumCircuit(2)
#qc.h(0)#NOT
qc.rx(np.pi/3,0) #NOT
qc.cx(0,1)
qc.draw("mpl")
```

```python
plot_bloch_multivector(qc)
```

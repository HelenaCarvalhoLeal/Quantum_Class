Calcule as energias dos três primeiros estados excitados (com cinco algarismos significativos) para o oscilador harmônico, pelo método do cachorro sacudindo o rabo (Problema 2.54). Para o primeiro (e terceiro) estado excitado você precisará estabelecer u[O] ==0, u'[O] = = 1.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Definindo a equação diferencial
def diff_eq(x, u, a):
    return [u[1], (x**2 - a**2) * u[0]]

# Valores iniciais
x_start, x_end = 0, 6
u0 = [1, 0]  # u(0) = 1, u'(0) = 0

# Resolvendo para diferentes valores de a
a_values = [4.99999, 6.9, 5.00001, 6.99999]
x = np.linspace(x_start, x_end, 1000)

# Criando os gráficos
plt.figure(figsize=(10, 10))

for i, a in enumerate(a_values, 1):
    sol = solve_ivp(diff_eq, [x_start, x_end], u0, args=(a,), t_eval=x, max_step=1e-2)
    
    plt.subplot(2, 2, i)
    plt.plot(sol.t, sol.y[0], label=f'a = {a}')
    plt.xlim([0, 6])
    plt.ylim([-1.5, 1.5])
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.title(f'Solução para a = {a}')
    plt.legend()

plt.tight_layout()
plt.show()
```
